```python
## Kick top, alpha = 1
## Multiple initial conditions for same k

import numpy as np
import random as rd
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
# %matplotlib notebook
plt.rcParams['figure.dpi'] = 80

fig = plt.figure()
# ax = Axes3D(fig)
# ax = fig.gca(projection='3d')
# ax = plt.axes(projection='3d')

x_array = [np.random.uniform(-1, 1) for l in np.zeros(200)]
y_array = np.array([])
i = 0
while i < 200:
    y_array = np.append(y_array, math.sqrt(1.-pow(x_array[i],2)))*np.random.uniform(-1, 1)
    i += 1
z_array = np.array([])
i = 0
while i < 200:
    z_array = np.append(z_array, math.sqrt(1.-pow(x_array[i],2)-pow(y_array[i],2)))*np.random.choice([-1,1])
    i += 1

k_array = np.array([0., 0.5, 0.97, 1.1, 1.2, 1.5, 2, 3, 5, 10])
# k_array = np.array([2])

alpha = 1.

for k in k_array:
    i = 0
    ax = plt.axes(projection='3d')
    while i < 200:
        x_n = x_array[i]
        y_n = y_array[i]
        z_n = z_array[i]
        
        x_ev = np.array([x_n])
        y_ev = np.array([y_n])
        z_ev = np.array([z_n])

        t = 0
        while t <= 10000:
            x_bar = x_n
            y_bar = y_n*math.cos(alpha)-z_n*math.sin(alpha)
            z_bar = y_n*math.sin(alpha)+z_n*math.cos(alpha)
            
            x_n1 = x_n*math.cos(k*z_bar)-y_bar*math.sin(k*z_bar)
            y_n1 = x_n*math.sin(k*z_bar)+y_bar*math.cos(k*z_bar)
            z_n1 = z_bar
            
            x_ev = np.append(x_ev,x_n1)
            y_ev = np.append(y_ev,y_n1)
            z_ev = np.append(z_ev,z_n1)
            
            x_n = x_n1
            y_n = y_n1
            z_n = z_n1
            
            t += 1
        
        ax.scatter(x_ev, y_ev, z_ev, s=0.5)
        i += 1

    plt.title('Perturbation Parameter: k = '+str(k))
    plt.tight_layout()
    plt.show()
```
